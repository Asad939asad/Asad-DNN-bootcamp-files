# Loading the file which contains all the rows on which we have to make preduictions

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
import joblib
# v=df
# Load the DataFrame
df = pd.read_csv('/content/drive/MyDrive/main_files/Combinedfile_filtered.csv')
for i,t in enumerate(df):
  print(i,t)
```

# Deleting Uncessary colums

```python
df.drop(['generation_W','load_W','timestamp','system_id','year','is_weekend','season','generation_to_capacity_ratio','time_since_start','time_standard','generation_W_rolling_mean_1h','peak_production','test_id','time_of_day'],axis=1,inplace=True)
```

```python
for i,t in enumerate(df):
  print(i,t)
```

```python
df['f1'] = (df['load_capacity'] * df['panels_capacity']) * df['month']
df['f2'] = (df['load_capacity'] * df['panels_capacity']) * df['hour']
df['f3'] = (df['load_capacity'] * df['panels_capacity']) * df['day_of_week']
```

```python
for i,t in enumerate(df):
  print(i,t)
```

# Scaling the input using saved scaler

```python
import joblib
sc=joblib.load('/content/minmax_scaler1forX (1).pkl')
data=sc.transform(df)
```

```python
X_train_reshaped = data.reshape((data.shape[0], 1, data.shape[1]))
```

# Mkaing predictions

```python
# Make predictions in batches rather than row by row
predictions = model.predict(X_train_reshaped)

# Extract predictions for generation_W and load_W
predicted_generation_W = predictions[:, 0]
predicted_load_W = predictions[:, 1]

# Add predictions to the DataFrame
df['predicted_generation_W'] = predicted_generation_W
df['predicted_load_W'] = predicted_load_W

# Save the updated DataFrame to a new CSV file
df.to_csv('/content/drive/MyDrive/main_files/predictions09p.csv', index=False)
```
