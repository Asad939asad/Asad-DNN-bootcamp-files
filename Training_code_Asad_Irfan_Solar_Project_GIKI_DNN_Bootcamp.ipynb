# Applying preprocessing and feature engineering to help the model to leaqrn patterns

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Extract year, month, day, hour, minute, and second
df['year'] = df['timestamp'].dt.year

df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df['hour'] = df['timestamp'].dt.hour
df['minute'] = df['timestamp'].dt.minute
df
```

```python
df = df.drop(columns=['timestamp'])
```

```python
df['day_of_week'] = pd.to_datetime(df[['year', 'month', 'day']]).dt.dayofweek
```

```python
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
def time_of_day(hour):
    if 5 <= hour < 12:
        return 0
    elif 12 <= hour < 17:
        return 1
    elif 17 <= hour < 21:
        return 2
    else:
        return 3
df['time_of_day'] = df['hour'].apply(time_of_day)
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 0
    elif month in [3, 4, 5]:
        return 1
    elif month in [6, 7, 8]:
        return 2
    else:
        return 3
df['season'] = df['month'].apply(get_season)
```

```python
df['load_to_capacity_ratio'] = df['load_W'] / df['load_capacity']
```

```python
df['generation_to_capacity_ratio'] = df['generation_W'] / df['panels_capacity']
```

```python
df['time_since_start'] = (pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute']]) - pd.to_datetime('2023-08-14')).dt.total_seconds()
```

```python
# Define a function for the time standard logic
def determine_time_standard(hour):
    return 1 if hour >= 12 else 0

# Apply the function to the 'hour' column
df['time_standard'] = df['hour'].apply(determine_time_standard)

```

```python
# Calculate rolling means
import numpy as np

df['generation_W_rolling_mean_1h'] = df['generation_W'].rolling(window=6).mean()
df['load_W_rolling_mean_1h'] = df['load_W'].rolling(window=6).mean()

# # Replace rolling means with 0 where original values are 0
# df['generation_W_rolling_mean_1h'] = np.where(
#     df['generation_W'] == 0,
#     0,
#     df['generation_W_rolling_mean_1h']
# )

# df['load_W_rolling_mean_1h'] = np.where(
#     df['load_W'] == 0,
#     0,
#     df['load_W_rolling_mean_1h']
# )
```

```python
# Define a function for the month-based production logic
def determine_peak_production(month):
    # Months 10 (October) to 4 (April) are non-peak, so return 0 for these months
    if 10 <= month <= 12 or 1 <= month <= 4:
        return 0
    # Otherwise, return 1 for peak production
    else:
        return 1

# Apply the function to the 'month' column
df['peak_production'] = df['month'].apply(determine_peak_production)
```
